```python
def t(time_str):
    try:
        if pd.isna(time_str) or time_str=="":
            return 0
        hours, minutes= map(int, time_str.split(':'))
        return hours+minutes/60
    except ValueError:
        return 0
df['hours worked decimal']= df['Hours Worked'].astype(str).apply(t)

df['half day']= ((df['hours worked decimal']>5 ) & (df['Status']=='P')).astype(int)

total_days= df.groupby('name')['half day'].sum()
print(total_days)
```

# Over Time

```python
# Ensure all entries in 'Hours Worked' are strings
df['Hours Worked'] = df['Hours Worked'].astype(str)

# Replace '0' values with '00:00'
df['Hours Worked'] = df['Hours Worked'].replace('0', '00:00')

# Convert 'Hours Worked' to timedelta
df['Hours Worked'] = pd.to_timedelta(df['Hours Worked'] + ':00', errors='coerce')
```

```python
# Define standard 8-hour workday as timedelta
standard_hours = pd.to_timedelta('08:00:00')

# Calculate overtime
df['Over Time'] = df['Hours Worked'].apply(lambda x: x - standard_hours if x > standard_hours else pd.Timedelta(0))

# Convert overtime to hours for easier reading
df['Over Time (hours)'] = df['Over Time'].dt.total_seconds() / 3600
```

```python
# Convert Over Time (hours) to HH:MM:SS format
def format_time(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    return f"{hours:02}:{minutes:02}:{secs:02}"

df['Over Time (formatted)'] = df['Over Time (hours)'].apply(lambda x: format_time(x * 3600))
```

```python
df.drop(columns=['Over Time (formatted)','Over Time (hours)'], inplace = True)
```

```python
total_hours= df.groupby('name')['Over Time'].sum()
```
